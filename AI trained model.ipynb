```python
disease_data = {
    "Flu": ["fever", "cough", "cold", "sore throat", "body pain", "chills", "fatigue"],
    "Covid-19": ["fever", "dry cough", "loss of taste", "loss of smell", "breathing difficulty", "tiredness"],
    "Dengue": ["high fever", "joint pain", "rash", "headache", "vomiting", "muscle pain", "bleeding gums"],
    "Malaria": ["fever", "chills", "sweating", "headache", "nausea", "shivering"],
    "Diabetes": ["frequent urination", "increased thirst", "weight loss", "fatigue", "blurred vision"],
    "Hypertension": ["headache", "chest pain", "dizziness", "blurred vision", "shortness of breath"],
    "Asthma": ["shortness of breath", "wheezing", "chest tightness", "coughing"],
    "Tuberculosis": ["persistent cough", "weight loss", "night sweats", "blood in sputum", "fever"],
    "Pneumonia": ["chest pain", "cough", "fever", "difficulty breathing", "fatigue"],
    "Typhoid": ["high fever", "weakness", "stomach pain", "diarrhea", "loss of appetite"],
    "Migraine": ["severe headache", "nausea", "sensitivity to light", "vomiting"],
    "Anemia": ["fatigue", "pale skin", "shortness of breath", "dizziness"],
    "Kidney Stone": ["severe pain", "blood in urine", "nausea", "pain while urinating"],
    "Appendicitis": ["lower right abdominal pain", "vomiting", "loss of appetite", "fever"],
    "Hepatitis": ["yellow skin", "dark urine", "fatigue", "nausea", "abdominal pain"],
    "Food Poisoning": ["vomiting", "diarrhea", "stomach cramps", "fever", "nausea"],
    "Bronchitis": ["cough", "mucus", "chest discomfort", "fatigue", "shortness of breath"],
    "Arthritis": ["joint pain", "swelling", "stiffness", "reduced motion"],
    "Sinusitis": ["face pain", "blocked nose", "headache", "runny nose"],
    "UTI": ["burning urination", "frequent urination", "lower abdominal pain", "cloudy urine"],
    "Heart Disease": ["chest pain", "shortness of breath", "sweating", "nausea", "fatigue"],
    "Stroke": ["face drooping", "arm weakness", "speech difficulty", "confusion"],
    "Depression": ["sadness", "loss of interest", "fatigue", "sleep problems", "hopelessness"],
    "Anxiety": ["nervousness", "restlessness", "rapid heartbeat", "sweating"],
    "Allergy": ["sneezing", "itchy eyes", "runny nose", "rash"],
    "Conjunctivitis": ["red eyes", "itching", "eye discharge", "watering eyes"],
    "Chickenpox": ["fever", "itchy rash", "blisters", "tiredness"],
    "Measles": ["fever", "rash", "runny nose", "red eyes", "cough"],
    "Obesity": ["weight gain", "fatigue", "breathing difficulty", "joint pain"]
}
```

```python
def predict_disease(user_symptoms):
    user_symptoms = user_symptoms.lower()
    max_match = 0
    predicted = "Unknown"

    for disease, symptoms in disease_data.items():
        match_count = 0
        for symptom in symptoms:
            if symptom in user_symptoms:
                match_count += 1

        if match_count > max_match:
            max_match = match_count
            predicted = disease

    if max_match == 0:
        return "Sorry, I couldn't identify the disease. Please consult a doctor."
    else:
        return f"Based on your symptoms, you may have: {predicted}. Please consult a doctor."
```

```python
import ipywidgets as widgets
from IPython.display import display

text = widgets.Text(
    description="Symptoms:",
    placeholder="Enter symptoms (fever, cough)"
)

button = widgets.Button(description="Predict")
output = widgets.Output()

def on_click(b):
    with output:
        output.clear_output()
        print(predict_disease(text.value))

button.on_click(on_click)

display(text, button, output)
```
